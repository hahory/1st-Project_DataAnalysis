```python
import pandas as pd
import xml.etree.ElementTree as ET
import os

# xml데이터 가져오기
tree = ET.parse('0617_01.xml')
root = tree.getroot()

# xml 구조 확인하기
for x in root:
    print(x.tag, x.attrib, x.text)
    for y in x:
        print("--",y.tag, y.attrib, y.text)
```

```python
# 추출 확인을 위해 print해봄

# image 태그에서 id와 name 추출하기
print(root[2].attrib['id'])
print(root[3].attrib['id'])
print(root[4].attrib['id'])
print(root[2].attrib['name'])
print(root[3].attrib['name'])
print(root[4].attrib['name'])

# box 태그에서 xtl, ytl 추출해보기
print(x[0].attrib['xtl'])
print(x[1].attrib['xtl'])
print(x[2].attrib['xtl'])
print(x[0].attrib['ytl'])
print(x[1].attrib['ytl'])
print(x[2].attrib['ytl'])
```

```python
# 필요한 속성들 추출 잘 되는 것 확인했고 for문으로 파일 내 전체 속성을 받아옴
# 좌표속성들(xtl,ytl,xbr,ybr) 추출하는 순서는 AI허브에서 제공하는 '09.인도 보행 영상.pdf'파일의 어노테이션 xml 파일 예시의 순서를 따랐음
rows=[]

for i in range(2,len(root)):
    id=root[i].attrib['id']
    name=root[i].attrib['name']
    x=root[i] #중요!!! 이전 파일에서 x를 선언해주지 않아서 x.attrib관련 수치들이 랜덤으로 들어왔었음
    for j in range(len(x)):
        label = x[j].attrib['label']
        xtl = x[j].attrib['xtl']
        ytl = x[j].attrib['ytl']
        xbr = x[j].attrib['xbr']
        ybr = x[j].attrib['ybr']
        rows.append({
            "id": id,
            "name": name,
            "label": label,
            "xtl": xtl,
            "ytl": ytl,
            "xbr": xbr,
            "ybr": ybr,
            })
```

```python
#rows 확인해보기
rows
```

```python
# 데이터프레임으로 변환했을 때 열(row) 수와 일치하는지 확인하기 위해 rows의 길이를 확인
len(rows)
```

```python
# 데이터프레임으로 보기
df = pd.DataFrame(rows)
df
# 데이터프레임의 열 수도 1684임을 확인
```

```python
# 각 이미지에 맞는 txt파일 만들기 위한 작업

# 여러 txt파일이 생성되므로 먼저 저장할 폴더 생성하기
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error:Creating directory.'+directory)
createFolder('../text_txt_files')


# # 반복문 작성 전 한 이미지파일에 대한 txt파일 만들기 테스트해본 것

# # name컬럼 값이 MP_SEL_000001.jpg인 행들만 추출
# df_image_1=(df['name']=='MP_SEL_000001.jpg')
# txt_1=df[df_image_1]
# # txt파일에 들어갈 컬럼들만 추출
# df_features=txt_1[['label','xtl','ytl','xbr','ybr']]
# # txt파일 이름 정하기(이미지이름과 동일하도록) 
# name=df['name'][0]
# file_name=name.split('.')[0]
# # txt파일로 저장
# df_features.to_csv('./txt_files/'+file_name+'.txt', sep=' ', index=False, header=None)
```

```python
# 위 셀에서 한 이미지파일에 대해 하나의 txt파일 만든 것을 반복문에 이용하여 모든 이미지파일에 적용하여 각각의 txt파일 만들기

# 파일 이름들이 저장되는 리스트 만들기
file_names=[]
for i in range(2, len(root)):
    name=root[i].attrib['name']
    file_names.append(name)

# 각 이미지파일에 해당하는 행들끼리 나눠주기
df_txt_files=[]
for i in range(len(file_names)):
    df_image=(df['name']==file_names[i])
    df_txt_file=df[df_image]
    df_txt_files.append(df_txt_file)

# 각 txt파일에 들어갈 컬럼들만 추출    
df_columns=[]
for i in range(len(df_txt_files)):
    df_column=df_txt_files[i][['label','xtl','ytl','xbr','ybr']]
    df_columns.append(df_column)

# 각 이미지이름과 동일하도록 txt파일 이름 정하고, 각 txt파일에 들어갈 컬럼들만 추출한 것으로 txt파일 만들기
for i in range(len(file_names)):
    file_name=file_names[i].split('.')[0]
    df_columns[i].to_csv('../text_txt_files/'+file_name+'.txt', sep=' ', index=False, header=None) 
```
